1. What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#load the califorina housing dataset
df=pd.read_excel('housing.xlsx')

#compute the median income
median_income=df['median_income'].median()

print('median_income :', median_income)

#plot the distribution of median incomes using a KDE plot
sns.kdeplot(data = df, x='median_income',fill=True)

## Add labels to the x-axis and y-axis, and a title to the plot
plt.title('The Distribution of Median Incomes using a KDE plot')
plt.xlabel('Median Income')
plt.ylabel('Density')

#show the plot
plt.show()
```


5. Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.



```python
import pandas as pd

#load california housing dataset
df= pd.read_excel('housing.xlsx')

#calculate the mean value of total_bedrooms in the orginal dataset
mean_total_bedroom = df['total_bedrooms'].mean()

#create the new dataset by filling missing values with the mean value of total_bedrooms
housing_filled= df.fillna(value={'total_bedrooms': mean_total_bedroom})

#print first 5 rows of the filled dataset
print(housing_filled.head())
```

10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.



```python
import numpy as np
import pandas as pd

#load the dataset
df=pd.read_excel('housing.xlsx')

#create a function for total_bedroom_size
def number_bedroom_size (num):
  if num <= 10:
    return 'small'

  elif num < 1000:
    return 'medium'

  else:
    return 'large'

df['total_bedroom_size']=df['total_bedrooms'].apply(number_bedroom_size)

#print(df[df['total_bedroom_size']== 'medium'])
print(df.head(20)) 
```
